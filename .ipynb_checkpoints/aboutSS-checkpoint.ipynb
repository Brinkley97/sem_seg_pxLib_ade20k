1. [d] SS differs from Instance Segmentation - see ex above
    1. SS : same item different regions = 1 category
        1. Trees, skies, buildings, etc are same color
    2. IS : same item different regions = 2 different categories
        1. Trees would be diff colors bc there are 5 diff instances of a tree, skies same bc 2 diff instances, etc
